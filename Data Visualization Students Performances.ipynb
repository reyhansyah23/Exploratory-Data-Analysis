```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
students = pd.read_csv('StudentsPerformance.xls')
students
```

Data diambil dari kaggle url: https://www.kaggle.com/spscientist/students-performance-in-exams

```python
students.dtypes
```

```python
students.rename(columns={"race/ethnicity": "race", "B": "c"}, inplace = True)
```

```python
students.drop(["lunch","test preparation course"], axis=1, inplace=True)
```

```python
students
```

```python
students['gender'] = students['gender'].astype('category')
students['race'] = students['race'].astype('category')
students['parental level of education'] = students['parental level of education'].astype('category')
```

```python
print(students.shape)
students.duplicated().sum()
```

```python
students.drop_duplicates(inplace=True)
print(students.duplicated().sum()) # no more duplicates
print(students.shape) # re-check by printing data size
```

```python
students_num = students.select_dtypes(include = ['float64', 'int64'])
students_num.head(10)
```

```python
from collections import Counter
```

```python
data = students['race']

proporsion = Counter(data)
values = [float(v) for v in proporsion.values()]
colors = ['r', 'g', 'b', 'y', 'c']
labels = proporsion.keys()
explode = (0, 0, 0, 0, 0)
plt.pie(values, colors=colors, labels= values, explode=explode, shadow=True)
plt.title('Kelas')
plt.legend(labels,loc='right')
plt.show()
```
